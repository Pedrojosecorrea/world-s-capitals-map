### 1) Import libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

### 2) Import data:

```python
df=pd.read_csv("concap.csv")
df.head()
```

### 3) Separated data:

```python
cap_lat=df['CapitalLatitude']
cap_lon=df['CapitalLongitude']
code_coun=df['CountryCode']
coun_name=df['CountryName']
conti_name=df['ContinentName']
capital_name=df['CapitalName']
```

### 4) Creating the map:

```python
#create map and initial point
map_cw = folium.Map(width=900,height=500, location=[cap_lat[0],cap_lon[0]], zoom_start=2)

#feature f1 is created 
f1=folium.FeatureGroup("Capitals")
```

### 5)Insert markers with world's capitals location:

```python
# the line type layer is created with the route matrix and added to the f1 characteristic 
for dat in range(len(cap_lat)):
    folium.Marker([cap_lat[dat],cap_lon[dat]], popup=str(coun_name[dat])+"("+str(code_coun[dat])+")"+"\n" + str(conti_name[dat]), tooltip=capital_name[dat],icon=folium.Icon(color="green", icon="info-sign")).add_to(f1)

#the feature (containing the line layer) is added to the map.   
f1.add_to(map_cw)

#add interactive icon to the map 
folium.LayerControl().add_to(map_cw)
map_cw.save('world_capitals.html')
```
